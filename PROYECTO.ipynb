```python


# === CONFIGURACI√ìN ===
ruta_base = "PROYECTON"   # carpeta con tus subcarpetas (fruver, mercado, etc.)
img_size = (70, 70)       # tama√±o uniforme para todas las im√°genes

# === 1. CARGA DE IM√ÅGENES Y ETIQUETAS ===
X = []  # vectores de im√°genes
y = []  # etiquetas (nombre de la carpeta)

for subcarpeta in os.listdir(ruta_base):
    ruta_sub = os.path.join(ruta_base, subcarpeta)
    if os.path.isdir(ruta_sub):
        for archivo in os.listdir(ruta_sub):
            if archivo.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')):
                ruta_img = os.path.join(ruta_sub, archivo)
                try:
                    img = Image.open(ruta_img).convert('L').resize(img_size)
                    X.append(np.array(img).flatten())   
                    y.append(subcarpeta)
                except Exception as e:
                    print(f"Error con {archivo}: {e}")

X = np.array(X)
y = np.array(y)

print(f" Im√°genes cargadas: {len(X)}")
print(f"Clases detectadas: {set(y)}")


modelo = {
    "imagenes": X,
    "etiquetas": y
}

def clasificar_imagen(ruta_img, modelo, img_size=(64, 64)):
    """
    Compara una nueva imagen con todas las del conjunto entrenado
    y devuelve la clase m√°s cercana (distancia euclidiana).
    """
    try:
        img = Image.open(ruta_img).convert('L').resize(img_size)
        vec = np.array(img).flatten()
    except:
        print(" No se pudo abrir la imagen.")
        return None

    # Calcular distancia euclidiana con todas las im√°genes del modelo
    distancias = np.linalg.norm(modelo["imagenes"] - vec, axis=1)
    indice_min = np.argmin(distancias)
    clase_predicha = modelo["etiquetas"][indice_min]

    print(f" La imagen '{os.path.basename(ruta_img)}' pertenece a la clase: '{clase_predicha}'")
    return clase_predicha

import shutil

def mover_a_carpeta(ruta_img, clase, ruta_base):
    destino = os.path.join(ruta_base, clase)
    os.makedirs(destino, exist_ok=True)
    shutil.move(ruta_img, os.path.join(destino, os.path.basename(ruta_img)))
    print(f" Imagen movida a: {destino}")

```
